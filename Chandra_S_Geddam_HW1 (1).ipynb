```python
# PROBLEM 1

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Modified URL for direct download from Google Drive
url = 'https://drive.google.com/uc?export=download&id=1r0fiSxVeEMXBBWxJ3K7ftI6Um-HFn8QR'
df = pd.read_csv(url)
df.head()

x1 = df.values[:, 0]
x2 = df.values[:, 1]
x3 = df.values[:, 2]
y = df.values[:, 3]

# Create X matrix with only x0 and x1 for linear regression with one feature
m = len(y)
x0 = np.ones(m)
X_1 = np.array([x0, x1]).T
X_2 = np.array([x0, x2]).T
X_3 = np.array([x0, x3]).T



def compute_cost(X, y, theta):
    """
    Compute cost for linear regression.

    Parameters:
    X : 2D array where each row is a training example and each column is a feature
        (including the intercept term)
    y : 1D array of the target values
    theta : 1D array of the parameters of the linear regression model

    Returns:
    J : float
        The cost function for the given parameters theta.
    """
    m = len(y)  # Number of training examples
    predictions = X.dot(theta)
    errors = np.subtract(predictions, y)
    sqErrors = errors**2
    J = (1 / (2 * m)) * np.sum(sqErrors)
    return J


def gradient_descent(X, y, theta, alpha, iterations):

    m = len(y)  # Number of training examples
    cost_history = np.zeros(iterations)

    for i in range(iterations):
        predictions = X.dot(theta)
        errors = np.subtract(predictions, y)
        sum_delta = (alpha / m) * X.transpose().dot(errors)
        theta -= sum_delta
        cost_history[i] = compute_cost(X, y, theta)

    return theta, cost_history

# Initialize theta for linear regression with one feature
theta_single = [0., 0.]
iterations = 1000
alpha = 0.05

# Using the X_single matrix as input to gradient_descent
theta_single, cost_history_x1 = gradient_descent(X_1, y, theta_single, alpha, iterations)
print('Final value of theta for single feature model =', theta_single)
print('cost_history for single feature model =', cost_history_x1)

# Scatter plot for the training data
plt.scatter(x1, y, color='red', marker='+', label='Training Data')

# Line plot for the linear regression model with single feature
plt.plot(x1, X_1.dot(theta_single), color='green', label='Linear Regression (x1 only)')

# Plot customizations
plt.rcParams["figure.figsize"] = (10, 6)
plt.grid(True)
plt.xlabel('x1')
plt.ylabel('y')
plt.title('Linear Regression Fit (x1 only)')
plt.legend()

# Show the plot
plt.show()

plt.plot(range(1, iterations + 1), cost_history_x1, color='blue')
plt.rcParams["figure.figsize"] = (10, 6)
plt.grid(True)

plt.xlabel('Number of iterations')
plt.ylabel('Cost (J)')
plt.title('Convergence of gradient descent')

# Show the plot
plt.show()
```

```python
X_2 = np.array([x0, x2]).T

theta_single, cost_history_x2 = gradient_descent(X_2, y, theta_single, alpha, iterations)
print('Final value of theta for single feature model =', theta_single)
print('cost_history for single feature model =', cost_history_x2)

# Scatter plot for the training data
plt.scatter(x2, y, color='red', marker='+', label='Training Data')

# Line plot for the linear regression model with single feature
plt.plot(x2, X_2.dot(theta_single), color='green', label='Linear Regression (x2 only)')

# Plot customizations
plt.rcParams["figure.figsize"] = (10, 6)
plt.grid(True)
plt.xlabel('x2')
plt.ylabel('y')
plt.title('Linear Regression Fit (x2 only)')
plt.legend()

# Show the plot
plt.show()

plt.plot(range(1, iterations + 1), cost_history_x2, color='blue')
plt.rcParams["figure.figsize"] = (10, 6)
plt.grid(True)

plt.xlabel('Number of iterations')
plt.ylabel('Cost (J)')
plt.title('Convergence of gradient descent')

# Show the plot
plt.show()
```

```python
X_3 = np.array([x0, x3]).T

theta_single, cost_history_x3 = gradient_descent(X_3, y, theta_single, alpha, iterations)
print('Final value of theta for single feature model =', theta_single)
print('cost_history for single feature model =', cost_history_x3)

# Scatter plot for the training data
plt.scatter(x3, y, color='red', marker='+', label='Training Data')

# Line plot for the linear regression model with single feature
plt.plot(x3, X_3.dot(theta_single), color='green', label='Linear Regression (x3 only)')

# Plot customizations
plt.rcParams["figure.figsize"] = (10, 6)
plt.grid(True)
plt.xlabel('x3')
plt.ylabel('y')
plt.title('Linear Regression Fit (x3 only)')
plt.legend()

# Show the plot
plt.show()

plt.plot(range(1, iterations + 1), cost_history_x3, color='blue')
plt.rcParams["figure.figsize"] = (10, 6)
plt.grid(True)

plt.xlabel('Number of iterations')
plt.ylabel('Cost (J)')
plt.title('Convergence of gradient descent')

# Show the plot
plt.show()
```

```python
# PROBLEM 2

# X matrix with bias + all features
X_A = np.array([x0, x1, x2, x3]).T

# Initialize theta for multi-feature model
theta = np.zeros(X_A.shape[1])

# Run gradient descent
theta, cost_history = gradient_descent(X_A, y, theta, 0.04, iterations) # LEARNING RATE = 0.04

print('Final value of theta for multi-feature model =', theta)
print('Final cost =', cost_history[-1])

# Predictions
y_pred = X_A.dot(theta)

# ---- Plot: Predicted vs Actual ----
plt.rcParams["figure.figsize"] = (10, 6)
plt.scatter(y, y_pred, color='blue', marker='o', label='Predicted vs Actual')
plt.plot([y.min(), y.max()], [y.min(), y.max()], color='red', linewidth=2, label='Ideal Fit (y = y_pred)')

plt.xlabel('Actual y')
plt.ylabel('Predicted y')
plt.title('Multiple Linear Regression Fit (x1, x2, x3)')
plt.legend()
plt.grid(True)
plt.show()

plt.plot(range(iterations), cost_history, color='green')
plt.xlabel('Iterations')
plt.ylabel('Cost J')
plt.title('Cost Function using Gradient Descent')
plt.grid(True)
plt.show()

```
